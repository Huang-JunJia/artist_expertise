```python
import requests
from bs4 import BeautifulSoup
```

```python
query = 'https://backend.deviantart.com/rss.xml?type=deviation&q=boost%3Apopular+in%3Atraditional%2F{}%2F{}&offset={}'
```

```python
offset = 0
categories = ['abstract', 'animals', 'fantasy', 'illustration', 'landscapes', 'macabre', 'other', 'people',
              'political', 'pop', 'portraits', 'psychedelic', 'scifi', 'spaceart', 'still', 'surreal']
```

```python
for art_type in ['drawings', 'paintings']:
    for category in categories:
        current_query = query.format(art_type, category, offset)
        r = requests.get(query)
        if r.status_code != requests.codes.ok:
            print ("Error processing request. Try again")
```

```python
from lxml import etree
doc = etree.fromstring(r.content)
```

```python
soup = BeautifulSoup(r.text,'xml')
```
